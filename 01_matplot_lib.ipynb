```python
import matplotlib.pyplot as plt
import pandas as pd
x = [1,2,3,4]
y = [1,9,6,5]
plt.plot(x,y,) # how to plot- (1,1),(2,9)..........
plt.grid() #for scale in the graph
plt.show()
```

***Pyplot API***

```python
# Unvirated numerical 
data ={
    "ID" : [1,2,3,4,5,6],
    "salary" :[5000,42342,11222,43434,3214,34323],
    "City" : ["DELHI","Mumbai","Goa","Goa","rajasthan","DELHI"]
}
df= pd.DataFrame(data)
```

```python
print(df.head())
```

```python
plt.plot(df["salary"], color="red", marker="o",linestyle=":",linewidth="1") #color-> for line color, marker for point of mark, line style for line stype dotted or etc, linewidth -> thickness of line 
plt.show()
```

**Histgogram**

```python
plt.hist(df["salary"])
plt.show()
```

** boxplot **

```python
plt.boxplot(df["salary"])
```

```python
# Unvirated strings
count=df["City"].value_counts()
plt.pie(count , labels=count.index ,autopct="%1.1f" , explode=[0,0.2,0.1,0]) # lable for name , autopct for percentage, explode for slice
plt.axis("equal") # to make complete circle
plt.show()
```

```python
plt.bar(count.index,count , color=["red","green","blue"])
plt.show()
```

```python
#  bvirated numerical-numberical 
plt.scatter(df["City"],df["ID"], color="blue")
```

```python
# line plot
sort=df["ID"].sort_values()
sort2=df["salary"].sort_values() # must sort them either its compelx result show if more values
plt.plot(df["ID"],df["salary"],marker="*")
```

```python
# bar chart
plt.bar(df["ID"], df["salary"], color="cyan", align="edge")
plt.show()
```
